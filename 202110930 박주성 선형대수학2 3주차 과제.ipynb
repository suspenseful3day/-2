```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([4, 3])

# Figure 객체 만들기
fig = plt.figure()

# 1x1개의 서브플롯을 추가, 첫번째 서브플롯 사용
ax = fig.add_subplot(1, 1, 1)

# 변수 a의 값을 좌표로 사용하여 점을 그린다
# x좌표, y좌표, s는 점의 크기
ax.scatter(a[0], a[1], s=30)

# 점 옆에 문자 a 표시, size는 문자의크기
ax.text(a[0] + 0.2, a[1] + 0.2, 'a', size=15)

# x축과 y축의 눈금 설정 (-5 ~ 5 )
ax.set_xticks(range(-5, 6))
ax.set_yticks(range(-5, 6))

# 그리드를 좌표축 아래에 표시
ax.grid()

# x축과 y축이 원점에서 만남
ax.set_axisbelow(True)

# x, y 축 눈금 비율 1:1
# 창 크기 변경해도 비율 유지
ax.set_aspect('equal', adjustable='box')

# 왼쪽 y축과 아래 x축을 (0,0)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
#위쪽, 오른쪽 축 제거
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 벡터 A와 B를 만듭니다.
A = np.array([1, -2])
B = np.array([4, 3])

# 1행 2열의 그래픽을 그릴 수 있는 환경을 만듦
f, ax = plt.subplots(1, 2)

ax[0].title.set_text('A')
ax[1].title.set_text('B')

# 각 화살표의 시작점 (0,0)을 나타냅니다.
ax[0].quiver(0, 0, A[0], A[1], angles='xy', scale_units='xy', scale=1)
ax[1].quiver(0, 0, B[0], B[1], angles='xy', scale_units='xy', scale=1)

# x축과 y축의 그래픽 범위
start_x = -1
end_x = 5
start_y = -3
end_y = 4

for i in range(2):
    ax[i].axis([start_x, end_x, start_y, end_y])
    ax[i].set_xticks(range(start_x, end_x))
    ax[i].set_yticks(range(start_y, end_y))
    ax[i].grid(True)
    ax[i].set_aspect('equal', adjustable='box')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 3차원 배열 A
A = np.array([4, 5, 6])
fig = plt.figure()

# 플롯 영역 추가: 3차원 프로젝션 사용
ax = fig.add_subplot(1, 1, 1, projection='3d')

# 벡터 (0, 0, 0)에서 A[0], A[1], A[2]까지의 화살표 그리기
ax.quiver(0, 0, 0, A[0], A[1], A[2], color='black', arrow_length_ratio=0.1)
ax.text(A[0], A[1], A[2], 'a', size=15)

# 그리드 생성
ax.set_xlim(0, 8)
ax.set_ylim(0, 8)
ax.set_zlim(0, 8)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.view_init(elev=20., azim=5)
ax.grid()
ax.set_axisbelow(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
a=np.array([5,3,0])
b=np.array([1,2,0])
#외적 계산
c=np.cross(a,b)
fig=plt.figure()
ax=fig.add_subplot(1,1,1,projection='3d')
#벡터a를 그린다
ax.quiver(0,0,0,a[0],a[1],a[2],color='black',arrow_length_ratio=0.1)
ax.text(a[0],a[1],a[2],'a',size=15)

#벡터b를 그린다
ax.quiver(0,0,0,b[0],b[1],b[2],color='red',arrow_length_ratio=0.1)
ax.text(b[0],b[1],b[2],'b',size=15)

#벡터a와b의 외적을 그린다
ax.quiver(0,0,0,c[0],c[1],c[2],color='blue',arrow_length_ratio=0.1)
ax.text(c[0],c[1],c[2],'c',size=15)
ax.set_xlim(0,7)
ax.set_ylim(0,7)
ax.set_zlim(0,8)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.view_init(elev=20,azim=5)
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('auto',adjustable='box')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
a=np.array([3,3,0])
b=np.array([6,6,0])
#외적 계산
c=np.cross(a,b)
fig=plt.figure()
ax=fig.add_subplot(1,1,1,projection='3d')
#벡터 a를 그린다
ax.quiver(0,0,0,a[0],a[1],a[2],color='black',arrow_length_ratio=0.1)
ax.text(a[0],a[1],a[2],'a',size=15)
#벡터 b를 그린다
ax.quiver(0,0,0,b[0],b[1],b[2],color='red',arrow_length_ratio=0.1)
ax.text(b[0],b[1],b[2],'b',size=15)
#벡터a와 벡터b의 외적을 그린다.
ax.quiver(0,0,0,c[0],c[1],c[2],color='blue',arrow_length_ratio=0.1)
ax.text(c[0],c[1],c[2],'c',size=15)
ax.set_xlim(0,7)
ax.set_ylim(0,7)
ax.set_zlim(0,8)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.view_init(elev=20,azim=5)
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('auto',adjustable='box')
plt.show()
```
